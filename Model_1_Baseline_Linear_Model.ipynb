# **Model 1 : Baseline Linear Model**

image.png

## Model 1 Explained: Baseline Linear Model

Model 1 implements a simple dynamic pricing strategy based on the historical occupancy of a parking spot. It uses a linear model to adjust the price based on the utilization (occupancy/capacity).

Here's a breakdown of the key steps:

1.  **Data Loading:** The code loads the parking data as a simulated stream using Pathway.
2.  **Utilization Calculation:** It calculates the utilization of each parking spot by dividing the occupancy by the capacity.
3.  **Stateful Reduction:** Pathway's `stateful_single` reducer is used to maintain the price for each parking spot over time.
4.  **Price Update:** The price is updated based on the previous price and the current utilization, using a simple linear formula: `new_price = last_price + ALPHA * previous_utilization`.
5.  **Price Clipping:** The price is clipped to stay within a defined range ([5, 20] in this case).
6.  **Joining Results:** The calculated prices are joined back to the original data to associate the price with each observation.

This model provides a basic baseline for dynamic pricing, demonstrating how Pathway can be used to process streaming data and maintain state for each entity (parking spot) in the stream. The subsequent models aim to improve upon this baseline by incorporating more features and sophisticated logic.

```python
# Pretend we loaded a CSV; swap in pw.io.kafka.read, pw.io.python.read, etc.
data: pw.Table = pw.io.csv.read("parking_stream.csv", schema=ParkingSchema)


# 2.  PRE‑CALC COLUMNS
# Utilization with safe divide
@pw.udf
def compute_utilization(occupancy: int, capacity: int) -> float:
    return 0.0 if capacity == 0 else occupancy / capacity

data = data.with_columns(
    Utilization = compute_utilization(data.Occupancy, data.Capacity)
)

# 3.  STATEFUL REDUCER  → replicates the ‘price_series’ loop

BASE_PRICE   = 10.0
ALPHA        = 0.5
LOWER_BOUND  = 5.0
UPPER_BOUND  = 20.0

@pw.reducers.stateful_single          # one row at a time, maintains state
def baseline_price(prev_price: float | None, util: float) -> float:
    """Pathway analogue of:
       new_price = last_price + ALPHA * previous_utilization"""
    if prev_price is None:                       # first row for the lot
        price = BASE_PRICE
    else:
        price = prev_price + ALPHA * util
    # clip to [5, 20] like pandas `.clip(lower=5, upper=20)`
    return max(LOWER_BOUND, min(UPPER_BOUND, price))

price_updates = (
    data.groupby(data.SystemCodeNumber)
        .reduce(
            BaselinePrice = baseline_price(data.Utilization),

            # latest in processing order
            Timestamp     = pw.reducers.latest(data.Timestamp),
            Occupancy     = pw.reducers.latest(data.Occupancy),
            Capacity      = pw.reducers.latest(data.Capacity),
        )
)

#  JOIN back to see every column together
#  (price_updates already contains the latest Occ/Cap)

price_updates = (
    data.groupby(data.SystemCodeNumber)
        .reduce(
            # manually propagate the key using a reducer
            SystemCodeNumber = pw.reducers.latest(data.SystemCodeNumber),
            BaselinePrice    = baseline_price(data.Utilization),
            Timestamp        = pw.reducers.latest(data.Timestamp),
            Occupancy        = pw.reducers.latest(data.Occupancy),
            Capacity         = pw.reducers.latest(data.Capacity),
        )
)

# Now you can safely reference SystemCodeNumber
result = price_updates.select(
    SystemCodeNumber = pw.this.SystemCodeNumber,
    Timestamp        = pw.this.Timestamp,
    Occupancy        = pw.this.Occupancy,
    Capacity         = pw.this.Capacity,
    BaselinePrice    = pw.this.BaselinePrice
)

```

```python
import pandas as pd
import numpy as np

np.random.seed(42)

# Parameters
num_lots = 3
time_steps = 18  # from 8:00 to 4:30 at 30-min intervals
base_price = 10.0
alpha = 0.5  # Controls slope of price change

records = []

for lot_id in range(num_lots):
    capacity = np.random.randint(50, 100)
    price = base_price
    for t in range(time_steps):
        timestamp = pd.Timestamp("2025-07-02 08:00") + pd.Timedelta(minutes=30 * t)
        occupancy = np.random.randint(0, capacity + 1)
        utilization = occupancy / capacity
        price += alpha * utilization  # No clamping!
        records.append({
            "ID": lot_id,
            "Timestamp": timestamp,
            "Occupancy": occupancy,
            "Capacity": capacity,
            "Price": price
        })

df_baseline = pd.DataFrame(records)
df_baseline.head(10)
```

```python
import datetime

BASE_PRICE = 10.0
ALPHA = 5.0  # Adjust this to control sensitivity

delta_window = (
    data_with_time.windowby(
        pw.this.t,                         # Event time column
        instance=pw.this.day,              # Partition per calendar day
        window=pw.temporal.tumbling(datetime.timedelta(days=1)),  # Daily window
        behavior=pw.temporal.exactly_once_behavior()
    )
    .reduce(
        t=pw.this._pw_window_end,
        occ_sum=pw.reducers.sum(pw.this.Occupancy),
        cap_sum=pw.reducers.sum(pw.this.Capacity),
        count=pw.reducers.count()
    )
    .with_columns(
        occ_avg=pw.this.occ_sum / pw.this.count,
        cap_avg=pw.this.cap_sum / pw.this.count
    )
    .with_columns(
        BASE_PRICE=BASE_PRICE + ALPHA * (pw.this.occ_avg / pw.this.cap_avg)
    )
)
```

```python
# Activate the Panel extension to enable interactive visualizations
pn.extension()

# Define a custom Bokeh plotting function that takes a data source (from Pathway) and returns a figure
def price_plotter(source):
    # Create a Bokeh figure with datetime x-axis
    fig = bokeh.plotting.figure(
        height=400,
        width=800,
        title="Pathway: Daily Parking Price",
        x_axis_type="datetime",  # Ensure time-based data is properly formatted on the x-axis
    )
    # Plot a line graph showing how the price evolves over time
    fig.line("t", "BASE_PRICE", source=source, line_width=2, color="navy")

    # Overlay red circles at each data point for better visibility
    fig.circle("t", "BASE_PRICE", source=source, size=6, color="red")

    return fig

# Use Pathway's built-in .plot() method to bind the data stream (delta_window) to the Bokeh plot
# - 'price_plotter' is the rendering function
# - 'sorting_col="t"' ensures the data is plotted in time order
viz = delta_window.plot(price_plotter, sorting_col="t")

# Create a Panel layout and make it servable as a web app
# This line enables the interactive plot to be displayed when the app is served
pn.Column(viz).servable()
```

# **Baseline Linear Pricing Model Report**

***Objective :***

To implement a simple and interpretable pricing model where the next price is determined linearly based on the previous price and the current occupancy level of an urban parking lot or a similar resource-constrained environment.


***Model Description :***

 Model 1: Baseline Linear Model

 Formula:

  Price(t+1) = Price(t) + alpha × (Occupancy / Capacity)

 Parameters:
 - Price(t): Current price at time t
 - Occupancy: Number of occupied units
 - Capacity: Total available units
 - alpha: Tuning parameter controlling price sensitivity to occupancy

***Key Assumptions :***

  - Occupancy directly affects price in a linear fashion.
  - Higher occupancy implies increased demand, thus higher pricing.
  - Model does not consider external variables such as time of day, location-specific dynamics, or competitor
 pricing.
  - The model assumes full observability of occupancy and capacity at each time step.

***Applications :***

  - Acts as a baseline or reference model in dynamic pricing projects.
  - Suitable for scenarios where simple, real-time adjustments are needed.- Can be used for:
  - Parking lots
  - Toll booths
  - Limited-seat events
  - Ride-sharing fare models (basic version)

***Limitations :***

 - Ignores competitor pricing, temporal trends, and customer behavior
 - Prone to overpricing under high but short
 - lived occupancy spikes- Cannot handle price elasticity or demand forecasting

***Conclusion : ***

  The Baseline Linear Pricing Model offers a foundational approach to dynamic pricing by directly linking future prices to current occupancy levels in a simple, interpretable manner. While it does not account for complex market dynamics or customer behavior, its linear structure provides a transparent and low-complexity mechanism to adjust prices based on real-time demand signals. This model is best suited as a benchmark or fallback strategy and lays the groundwork for more sophisticated models that incorporate demand forecasting, competitor pricing, and temporal factors.
  


# **Model 2 : Demand-Based Price Function**

image.png

```python
# Fill any missing values
df = df.fillna(0)

# Assign vehicle type weights
vehicle_weights = {
    'car': 1.0,
    'bike': 0.5,
    'truck': 1.5
}

TrafficConditionNearby_weights={
    'low': 0.5,
    'average': 1.0,
    'high': 1.5
}

df['VehicleWeight'] = df['VehicleType'].map(vehicle_weights).fillna(1.0)  # Default weight = 1.0
df['TrafficConditionNearby'] = df['TrafficConditionNearby'].map(TrafficConditionNearby_weights).fillna(1.0)  # Default weight = 1.0

# Safety check
df['Capacity'] = df['Capacity'].replace(0, np.nan)  # avoid div-by-zero
```

```python
# Coefficients
α = 2.0     # Occupancy weight
β = 0.5     # Queue length weight
γ = 1.0     # Traffic penalty
δ = 2.0     # Special day bonus
ε = 1.5     # Vehicle type weight
λ = 0.8     # Price sensitivity

# Compute demand score
df['Utilization'] = df['Occupancy'] / df['Capacity']
df['DemandScore'] = (
    α * df['Utilization'] +
    β * df['QueueLength'] -
    γ * df['TrafficConditionNearby'] +
    δ * df['IsSpecialDay'] +
    ε * df['VehicleWeight']
)
```

```python
# Normalize demand to [0, 1]
min_demand = df['DemandScore'].min()
max_demand = df['DemandScore'].max()

df['NormalizedDemand'] = (df['DemandScore'] - min_demand) / (max_demand - min_demand)
```

```python
BASE_PRICE = 10

df['Model2_Price'] = BASE_PRICE * (1 + λ * df['NormalizedDemand'])

# Clip to price bounds
df['Model2_Price'] = df['Model2_Price'].clip(lower=5, upper=20)

# Preview
df[['SystemCodeNumber', 'Timestamp', 'Occupancy', 'QueueLength', 'TrafficConditionNearby', 'IsSpecialDay', 'VehicleType', 'Model2_Price']].head()
```
